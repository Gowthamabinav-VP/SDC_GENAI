```python
# Step 1: Install necessary libraries (Only needed if not already installed)
!pip install transformers torch --quiet

# Step 2: Import the required libraries
from transformers import pipeline

# Step 3: Create a summarizer using Hugging Face's pre-trained model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Step 4: Example text to summarize
text = """
The discovery of the structure of DNA in 1953 by Watson and Crick is one of the greatest scientific accomplishments in history.
This breakthrough revolutionized the fields of genetics and molecular biology, leading to profound advancements in the understanding of heredity, disease, and biotechnology.
DNA, or deoxyribonucleic acid, is a molecule that carries genetic instructions for the development, functioning, and reproduction of all living organisms.
Its double helix structure allows for the efficient storage and transmission of genetic information across generations.
The implications of this discovery have had far-reaching effects, including the mapping of the human genome and the development of new treatments for various diseases, including cancer.
"""

# Step 5: Summarize the text
summary = summarizer(text, max_length=100, min_length=50, do_sample=False)

# Step 6: Output the summary
print("Original Text:")
print(text)
print("\nSummarized Text:")
print(summary[0]['summary_text'])
```
